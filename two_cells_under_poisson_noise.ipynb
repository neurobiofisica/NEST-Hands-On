```python
import os
import pylab as pl
import numpy as np
import nest
```

```python
## simulation parameters
T = 2000.                # simulation time (ms)
dt = 0.05                 # simulation resolution (ms)


## neuron parameters
Vth = 10.                # spike threshold (mV)
Vrest = 0.               # resting potential (mV)
Vreset = 0.              # reset potential (mv)
Tref = 10.               # Refractory period (ms)
```

```python
## set up and run simulation
nest.ResetKernel()
nest.SetStatus([0],{
    'resolution': dt,       ## set simulation resolution
    'print_time': False,    ## enable printing of simulation progress
    })
```

```python
neuron_params = {'I_e': 0., 'E_L': Vrest, 'V_th': Vth, 'V_reset': Vreset, 't_ref' : Tref, 'tau_m' : 20.}
neuron1 = nest.Create('iaf_psc_delta', params=neuron_params)

neuron_params = {'I_e': 0., 'E_L': Vrest, 'V_th': Vth, 'V_reset': Vreset, 't_ref' : Tref, 'tau_m' : 20.}
neuron2 = nest.Create('iaf_psc_delta', params=neuron_params)
```

```python
voltmeter = nest.Create('voltmeter')              # Setting up the voltmeter
nest.SetStatus(voltmeter, {'interval' : 0.5})   # Setting the voltmeter time step

nest.Connect(voltmeter, neuron1)
nest.Connect(voltmeter, neuron2)
```

```python
nest.SetDefaults("poisson_generator", {"rate": 10.})
noise = nest.Create("poisson_generator")
nest.Connect(noise, neuron1)
nest.Connect(noise, neuron2)
```

```python
sd1 = nest.Create('spike_detector')
nest.ConvergentConnect(neuron1, sd1)
sd2 = nest.Create('spike_detector')
nest.ConvergentConnect(neuron2, sd2)
```

```python
def run(r, noise, n1, n2, T, voltmeter):
    
    print
    print '\x1b[1m\x1b[34m --- \x1b[0m'

    # Print informations about where the program is...
    print "\033[94m Starting simulation with r = " + str(r) + "\x1b[0m"

    ## Restarting everything
    # Restarting the simulation time
    nest.SetStatus([0],{
            'time' : 0.0
            })

    # set random initial membrane potentials
    nest.SetStatus(n1, { 'V_m': Vrest })
    nest.SetStatus(n2, { 'V_m': Vrest })

    # Updating the noise rate
    nest.SetStatus( noise, { "rate" : r })

    # Running the simulation after all
    nest.Simulate(T)
    
    # read out recording time and voltage from voltmeter
    times=nest.GetStatus(voltmeter)[0]['events']['times']
    voltage=nest.GetStatus(voltmeter)[0]['events']['V_m']

    ## Now we separate the results of neuron 1 and neuron 2
    times1 = [times[i] for i in range(len(times)) if i % 2 == 0]
    times2 = [times[i] for i in range(len(times)) if i % 2 == 1]
    voltage1 = [voltage[i] for i in range(len(voltage)) if i % 2 == 0]
    voltage2 = [voltage[i] for i in range(len(voltage)) if i % 2 == 1]
    
    # Number of spikes fired by neuron1
    print '#spikes fired by neuron1: ', nest.GetStatus(sd1)[0]['n_events']
    # Number of spikes fired by neuron1
    print '#spikes fired by neuron2: ', nest.GetStatus(sd2)[0]['n_events']
    
    ## Plotting both graphs!
    fig1 = pl.figure(1, figsize=(10,10))
    fig1.clf()
    p1 = fig1.add_subplot(2,1,1)
    p2 = fig1.add_subplot(2,1,2)
    
    p1.plot(times1, voltage1, color='Black', linewidth=1.5)
    p1.set_xlabel('time (ms)')
    p1.set_ylabel('Neuron1 - V (mV)')
    
    p2.plot(times2, voltage2, color='Black', linewidth=1.5)
    p2.set_xlabel('time (ms)')
    p2.set_ylabel('Neuron 2 - V (mV)')
    
    pl.show()
    
    return
```

```python
run(10.0, noise, neuron1, neuron2, T, voltmeter)
```
